5 a .We want to consider only numeric data ,
 hence we will create a new dataframe which consists of only numeric data.

```python
df_numeric= df.select_dtypes(include=['float64', 'int64'])
df_numeric
```

5 b .Now, we shall remove data for '2012' , 'Base Year' and ' Unnamed 16' since they are essentially of no use to us

```python
df_numeric=df_numeric.drop(['2012','Unnamed: 16','Base Year'], axis =1) #axis=1 refers to the column
df_numeric
```

Now , the new dataframe contains:

```python
df_numeric.info()
```

6.Dividing the data into input and output:

Output - 2021(we shall predict the growth for year 2021)

Input - All the columns except 2021( we shall use the data from year 2013 to 2020 to predict the growth for 2021 for each country specified by their respective country code)

```python
x= df_numeric.iloc[:,0:9].values    # (input)x contains all rows and columns from index 0 till 8
x
```

```python
y= df_numeric.iloc[:,9].values  # (output)y = column 2021
y
```

7.Train and Test Variables

Import library train_test_split from package sklearn.model_selection

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,random_state=0)
```

x is input data.
Here , we see that 75% of x goes to input training data and 25% of x goes to input testing data.

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

y is output data.
Here , we see that 75% of y goes to output training data and 25% of y goes to output testing data.

```python
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```

8.Scaling and Normalization
It is done only for input data, in order to remove irregularity in input data(due to various decimals).

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

9.Run a Classifier/Regressor/Clusterer

Here, we perform linear regression.

We shall use LinerRegression (library) from sklearn.linear_model (package).

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

10.Model Fitting

Values of x_train and y_train are put inside the Linear Regression Library

```python
model.fit(x_train, y_train)  
```

11.Predict the output

By taking the input testing data, we predict the output.

y_pred will have the predicted values

```python
y_pred = model.predict(x_test) 
y_pred  
```
